```python
data_na = data.dropna()
```

```python
dummies = pd.get_dummies(data_na['ocean_proximity'], dtype=int)
```

```python
data_na = data_na.join(dummies)
```

```python
data_na.drop('ocean_proximity', axis=1, inplace=True)
```
