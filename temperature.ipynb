```python
# Vamos criar um modelo de regress√£o linear que converte C para F
import pandas as pd
data = pd.read_csv('temperatures.csv')
```

```python
import seaborn as sb

sb.scatterplot(x='celcius', y='fahrenheit', data=data, 
               hue='fahrenheit', 
               palette='coolwarm'
               )
```

```python
X = data['celcius']
y = data['fahrenheit']
```

```python
X
```

```python
y
```

```python
X_processed = X.values.reshape(-1, 1)
```

```python
y_processed = y.values.reshape(-1, 1)
```

```python
# Model creation

from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(X_processed, y_processed)
```

```python
model.predict([[123]])
```

```python
model.score(X_processed, y_processed)
```
